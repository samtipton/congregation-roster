```python
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
```

```python
a = pd.read_csv('../data/previous-assignments.csv', index_col=0)
a_drop_rounds = a.drop(['Rounds'], axis=1)
```

```python
avg_normalized = pd.DataFrame(index=a.index)

for person in a.index:
    for task in a_drop_rounds.columns:
        avg_normalized[f"{task}"] = a[task] / a[task].sum()
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
seaborn.heatmap(avg_normalized, robust=True)
```

```python
fig, ax = plt.subplots(figsize=(30, 10))
plt.xticks(rotation=45)
sp = seaborn.scatterplot(a_drop_rounds.drop(["lesson", "announcements", "usher", "alt_usher", "security"], axis=1))
```
