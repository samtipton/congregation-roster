```python
a = pd.read_csv('previous-assignments-ipynb.csv', index_col=0)
a_drop_rounds = a.drop(['Rounds'], axis=1)
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
seaborn.heatmap(avg_normalized, robust=True)
```
